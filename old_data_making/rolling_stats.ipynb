```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data/bbb_reindexed.csv", low_memory=False)
```

```python
valid_teams = ['Australia', 'West Indies', 'Pakistan', 'Bangladesh', 'South Africa', 'England', 'New Zealand', 'India', 'Sri Lanka']

df = df[df['bat_team'].isin(valid_teams) & df['bowl_team'].isin(valid_teams)]
```

```python
team_df = df.sort_values(by='bat_team')
team_df
```

```python
wickets = (
    df[df['wicket_type'].notna()]
    .groupby(['match_id', 'bowl_team'])
    .size()
    .reset_index(name='wickets_taken')
)

team_rows = []

# Group by match
for match_id, match_df in df.groupby('match_id'):
    date = match_df['match_date'].iloc[0]

    # Split innings
    inns_1 = match_df[match_df['innings'] == 1]
    inns_2 = match_df[match_df['innings'] == 2]
    
    if inns_1.empty or inns_2.empty:
        continue 

    # Teams
    team1 = inns_1['bat_team'].iloc[-1]
    team2 = inns_2['bat_team'].iloc[-1]

    # Total runs in last row of each innings (accumulated column)
    runs1 = inns_1['team_total_runs'].iloc[-1]
    runs2 = inns_2['team_total_runs'].iloc[-1]

    # Wickets taken (by bowling team)
    wkts1 = wickets.query("match_id == @match_id and bowl_team == @team1")['wickets_taken']
    wkts2 = wickets.query("match_id == @match_id and bowl_team == @team2")['wickets_taken']
    wkts1 = int(wkts1.iloc[0]) if not wkts1.empty else 0
    wkts2 = int(wkts2.iloc[0]) if not wkts2.empty else 0

    # Determine winner based on target (usually only in inns 2)
    target = inns_2['target'].dropna().unique()
    target = target[0] if len(target) > 0 else None
    winner = team2 if target is not None and runs2 >= target else team1

    # Add team1 row
    team_rows.append({
        'match_id': match_id,
        'match_date': date,
        'team': team1,
        'runs_scored': runs1,
        'won': int(team1 == winner),
        'wickets_taken': wkts1
    })

    # Add team2 row
    team_rows.append({
        'match_id': match_id,
        'match_date': date,
        'team': team2,
        'runs_scored': runs2,
        'won': int(team2 == winner),
        'wickets_taken': wkts2
    })

# Create match-level summary DataFrame
match_team_stats = pd.DataFrame(team_rows)
match_team_stats = match_team_stats.sort_values(['match_id', 'team']).reset_index(drop=True)

# Preview
print(match_team_stats.head())
```

```python
# Sort by team and date
match_team_stats = match_team_stats.sort_values(['team', 'match_date']).reset_index(drop=True)

# Group by team
grouped = match_team_stats.groupby('team')

# Rolling calculations (shift to exclude current match)
match_team_stats['win_rate_last_5'] = (
    grouped['won']
    .rolling(window=5, min_periods=1)
    .mean()
    .shift(1)
    .reset_index(level=0, drop=True)
)

match_team_stats['avg_runs_last_5'] = (
    grouped['runs_scored']
    .rolling(window=5, min_periods=1)
    .mean()
    .shift(1)
    .reset_index(level=0, drop=True)
)

match_team_stats['avg_wkts_last_5'] = (
    grouped['wickets_taken']
    .rolling(window=5, min_periods=1)
    .mean()
    .shift(1)
    .reset_index(level=0, drop=True)
)

# Fill NaNs for teams' first few matches
match_team_stats[['win_rate_last_5', 'avg_runs_last_5', 'avg_wkts_last_5']] = (
    match_team_stats[['win_rate_last_5', 'avg_runs_last_5', 'avg_wkts_last_5']].fillna(0)
)
```

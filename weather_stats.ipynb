```python
# 1. Build innings summary & pivot to find winners (same as before)
innings_summary = (
    df[['match_id','inns','team_bat','inns_runs','target','date']]
      .drop_duplicates(subset=['p_match','inns','team_bat'])
      .rename(columns={'team_bat':'team', 'inns_runs':'runs_scored'})
)

match_pivot = (
    innings_summary
    .pivot(index=['p_match','date','target'], columns='inns', values=['team','runs_scored'])
)
match_pivot.columns = [f"{stat}_{inns}" for stat, inns in match_pivot.columns]
match_pivot = match_pivot.reset_index()

match_pivot['winner'] = match_pivot.apply(
    lambda r: r['team_2'] if r['runs_scored_2'] >= r['target'] else r['team_1'],
    axis=1
)

# 2. Melt back to one row per team
rows = []
for _, r in match_pivot.iterrows():
    for innings in [1,2]:
        rows.append({
            'p_match':     r['p_match'],
            'date':        r['date'],
            'team':        r[f'team_{innings}'],
            'runs_scored': r[f'runs_scored_{innings}'],
            'won':         int(r[f'team_{innings}'] == r['winner'])
        })
match_stats = pd.DataFrame(rows)

# 3. Compute wickets taken per match per team
wickets = (
    df[df['dismissal'].notna()]
      .groupby(['p_match','team_bowl'], as_index=False)
      .size()
      .reset_index(names='wickets_taken')
      .rename(columns={'team_bowl':'team'})
)
match_stats = match_stats.merge(wickets, on=['p_match','team'], how='left').fillna({'wickets_taken':0})

# 4. Sort by team and date
match_stats = match_stats.sort_values(['team','date'])

# 5. Compute rolling features via groupby.apply
n = 5
def rolling_feats(group):
    rolled = (
        group[['won','runs_scored','wickets_taken']]
        .rolling(window=n, min_periods=1)
        .mean()
        .shift(1)  # exclude current match
        .rename(columns={
            'won':'win_rate_last_n',
            'runs_scored':'avg_runs_last_n',
            'wickets_taken':'avg_wkts_last_n'
        })
    )
    return pd.concat([group.reset_index(drop=True), rolled.reset_index(drop=True)], axis=1)

match_stats = match_stats.groupby('team', group_keys=False).apply(rolling_feats)

# 6. Merge back onto ball-by-ball for batting and bowling
for side in ['bat','bowl']:
    df = df.merge(
        match_stats[['p_match','team','win_rate_last_n','avg_runs_last_n','avg_wkts_last_n']],
        left_on=['p_match',f'team_{side}'],
        right_on=['p_match','team'],
        how='left'
    ).rename(columns={
        'win_rate_last_n': f'{side}_win_rate_last_n',
        'avg_runs_last_n': f'{side}_avg_runs_last_n',
        'avg_wkts_last_n': f'{side}_avg_wkts_last_n'
    }).drop(columns=['team'])

# Now df has the rolling metrics for both batting and bowling sides.
```

```python
# FUCK IT NEW APPROACH
# Precompute dismissals for each match and bowling team
wickets = (
    df[df['dismissal'].notna()]
    .groupby(['p_match', 'team_bowl'])
    .size()
    .reset_index(name='wickets_taken')
)

team_rows = []

# Group by match
for match_id, match_df in df.groupby('p_match'):
    # Basic match-level info
    date = match_df['date'].iloc[0]

    # Ensure both innings exist
    inns_1 = match_df[match_df['inns'] == 1]
    inns_2 = match_df[match_df['inns'] == 2]
    if inns_1.empty or inns_2.empty:
        continue  # skip incomplete matches

    # Teams and runs
    team1 = inns_1['team_bat'].iloc[-1]
    runs1 = inns_1['inns_runs'].iloc[-1]

    team2 = inns_2['team_bat'].iloc[-1]
    runs2 = inns_2['inns_runs'].iloc[-1]

    # wickets
    wkts1 = wickets.query("p_match == @match_id and team_bowl == @team1")['wickets_taken']
    wkts2 = wickets.query("p_match == @match_id and team_bowl == @team2")['wickets_taken']
    wkts1 = int(wkts1.iloc[0]) if not wkts1.empty else 0
    wkts2 = int(wkts2.iloc[0]) if not wkts2.empty else 0
    
    # Target-based win logic
    target = inns_2['target'].dropna().unique()
    target = target[0] if len(target) > 0 else None
    winner = team2 if target is not None and runs2 >= target else team1

    # Add two rows: one for each team
    team_rows.append({
        'p_match': match_id,
        'date': date,
        'team': team1,
        'runs_scored': runs1,
        'won': int(team1 == winner),
        'wickets_taken': wkts1
    })

    team_rows.append({
        'p_match': match_id,
        'date': date,
        'team': team2,
        'runs_scored': runs2,
        'won': int(team2 == winner),
        'wickets_taken': wkts2
    })



# Create match-level team stats DataFrame
match_team_stats = pd.DataFrame(team_rows)

# Sort by match and reset index
match_team_stats = match_team_stats.sort_values(['p_match', 'team']).reset_index(drop=True)
```
